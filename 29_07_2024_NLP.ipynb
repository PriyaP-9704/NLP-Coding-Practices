```python
# steps to be followed for NLP pre processing

# step 1 - convert upper to lower case
# step 2 - remove emoji's,special char using packages()
# step 3 - remove meaningless words (stop words)
# step 4 - get the root words (lemmatization)
# step 5 - part of speech tagging (pos_tag)
# step 6 - chunking(joining)based on pos_tag
```

```python
# natural lang
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop_word = stopwords.words('english')
```

```python
stop_word
```

```python
# new list containing no stop words
final_data = []

for word in data['Review-Body'][1].split():
  if word.lower() not in stop_word:
    final_data.append(word)
```

```python
x= data['Review-Body'][1]
x
```

```python
final_data
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
new_root_word = []

for word in final_data:
  doc = nlp(word)
  new_root_word.append([token.lemma_ for token in doc])
```

```python
new_root_word
```

```python
nltk.download('averaged_perceptron_tagger')
```

```python
avg_per_tag = []
for word in new_root_word:
  avg_per_tag.append(nltk.pos_tag(word))
```

```python
avg_per_tag
```

```python
chnk = []
for word in avg_per_tag:
  chnk.append(nltk.ne_chunk(word))
```

```python
from sklearn.manifold import TSNE #TSNE is a 3D visualization tool stands for t-Distributed Stochastic Neighbor Embedding
import plotly.express as px
import gensim.downloader as api
import numpy as np

# Load pre-trained Word2Vec model
model = api.load("glove-wiki-gigaword-300")

# Sample data
texts = [
    "I am a non vegeterian",
    "I have a practice to read books",
"Non vegeterian is the best choice for human",
"I love comedy movies",
"Crime novel is the best always",
"Comedian always play a vital role in cinema"]

def text_to_vector(text, model):
    words = text.split()
    word_vectors = [model[word] for word in words if word in model]
    if len(word_vectors) == 0:
        return np.zeros(model.vector_size)
    return np.mean(word_vectors, axis=0)
vectors = np.array([text_to_vector(text, model) for text in texts])

# Apply t-SNE to reduce dimensions to 3D
tsne = TSNE(n_components=3, perplexity=2, random_state=42)
X_3d = tsne.fit_transform(vectors)

# Create a DataFrame for Plotly
import pandas as pd
df = pd.DataFrame(X_3d, columns=['x', 'y', 'z'])
df['text'] = texts

# Plot with Plotly
fig = px.scatter_3d(df, x='x', y='y', z='z', text='text')
fig.update_traces(marker=dict(size=5), selector=dict(mode='markers'))
fig.show()
```
